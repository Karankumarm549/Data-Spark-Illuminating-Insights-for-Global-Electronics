```python
dup=product.duplicated().sum()
```

```python
product['Unit Cost USD'] = product['Unit Cost USD'].replace({'\$': ''}, regex=True)
product['Unit Price USD'] = product['Unit Price USD'].replace({'\$': ''}, regex=True)
product['Unit Cost USD']=pd.to_numeric(product['Unit Cost USD'],errors='coerce')
product['Unit Price USD']=pd.to_numeric(product['Unit Price USD'],errors='coerce')
```

```python
product['Unit Cost USD'].fillna(product['Unit Cost USD'].mean(),inplace=True)
product['Unit Price USD'].fillna(product['Unit Price USD'].mean(),inplace=True)
```

```python
product.info()
```

```python
product.to_csv('Products_C.csv',index=False)
```

```python
# Insert data
for index, row in product.iterrows():
    cursor.execute("""
    INSERT INTO product (ProductKey,Product_Name,Brand,Color,Unit_Cost_USD,Unit_Price_USD,SubcategoryKey,Sub_Category,CategoryKey,Category)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))

# Commit the transaction
conn.commit()

# Close the connection
cursor.close()
conn.close()

print("Data added to MySQL database successfully!")
```
