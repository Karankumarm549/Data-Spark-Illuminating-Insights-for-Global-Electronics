```python
# Establish a connection to your MySQL database
conn = mysql.connector.connect(
    host='localhost',       # your database host
    user='root',   # your MySQL username
    password='@Karan549', # your MySQL password
    database='dataspark' # the database where you want to insert the data
)
```

```python
	CustomerKey	Gender	Name	City	State Code	State	Zip Code	Country	Continent	Birthday
```

```python
# Create a cursor object
cursor = conn.cursor()

# Create the Products table if it doesn't exist
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers (
    CustomerKey INT PRIMARY KEY,
    Gender VARCHAR(10),
    Name VARCHAR(100),
    City VARCHAR(100),
    State_Code VARCHAR(30),
    State VARCHAR(30),
    Zip_Code VARCHAR(20),
    Country VARCHAR(30),
    Continent VARCHAR(30),
    Birthday DATE
);
""")
```

```python
# Insert data into the table row by row
for index, row in CUS.iterrows():
    cursor.execute("""
    INSERT INTO customers (CustomerKey,Gender,Name,City,State_Code,State,Zip_Code,Country,Continent,Birthday)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))

# Commit the transaction
conn.commit()

# Close the connection
cursor.close()
conn.close()

print("Data added to MySQL database successfully!")
```
