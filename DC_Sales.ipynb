```python
import mysql.connector
```

```python
sales_data = pd.read_csv('Sales_C.csv')
```

```python
# Establish a connection to your MySQL database
conn = mysql.connector.connect(
    host='localhost',     
    user='root',   
    password='@Karan549', 
    database='dataspark'  
)

# Create a cursor object
cursor = conn.cursor()
```

```python
#Order Number	Line Item	Order Date	Delivery Date	CustomerKey	StoreKey	ProductKey	Quantity	Currency Code
```

```python
# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Sales (
    Order_Number INT,
    Line_Item INT,
    Order_Date DATE,
    Delivery_Date DATE,
    CustomerKey INT,
    StoreKey INT,
    ProductKey INT,
    Quantity INT,
    Currency_Code VARCHAR(10)
);
""")
```

```python
# Insert data
for index, row in sales_data.iterrows():
    cursor.execute("""
    INSERT INTO Sales (Order_Number,Line_Item,Order_Date,Delivery_Date,	CustomerKey,	StoreKey,	ProductKey,	Quantity,	Currency_Code)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))

# Commit the transaction
conn.commit()

# Close the connection
cursor.close()
conn.close()

print("Data added to MySQL database successfully!")
```
