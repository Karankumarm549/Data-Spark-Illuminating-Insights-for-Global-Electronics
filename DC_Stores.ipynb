```python
import mysql.connector
```

```python
Store = pd.read_csv('Stores_C.csv')
```

```python
#Establish a connection to your MySQL database
conn = mysql.connector.connect(
    host='localhost',       # your database host
    user='root',   # your MySQL username
    password='@Karan549', # your MySQL password
    database='dataspark' # the database where you want to insert the data
)
```

```python
# Create a cursor object
cursor = conn.cursor()

# Create the Products table if it doesn't exist
cursor.execute("""
CREATE TABLE IF NOT EXISTS store (
    StoreKey INT PRIMARY KEY,
    Country VARCHAR(50),
    State VARCHAR(100),
    Squre_Meter INT,
    Open_Date DATE
);
""")
```

```python
#Insert data into the table row by row
for index, row in Store.iterrows():
    cursor.execute("""
    INSERT INTO store (StoreKey,Country,State,Squre_Meter,Open_Date)
    VALUES (%s, %s, %s, %s, %s)
    """, tuple(row))

# Commit the transaction
conn.commit()

# Close the connection
cursor.close()
conn.close()

print("Data added to MySQL database successfully!")
```
